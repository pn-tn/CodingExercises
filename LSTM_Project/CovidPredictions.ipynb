```python
# split into a sequence
from numpy import array

def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        end_ix = i + n_steps
        
        if end_ix > len(sequence)-1:
            break
            
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)
```

```python
# define the n_steps and n_features

n_steps = 5
n_features = 1
```

```python
# split the data at n_steps

X_tr, y_tr = split_sequence(training_array,n_steps)
X_te, y_te = split_sequence(testing_array,n_steps)
X_va, y_va = split_sequence(validation_array,n_steps)

# scaled version of the split
X_tr_s, y_tr_s = split_sequence(scaled_training_array,n_steps)
X_te_s, y_te_s = split_sequence(scaled_testing_array,n_steps)
X_va_s, y_va_s = split_sequence(scaled_validation_array,n_steps)

print(X_tr.shape)
print(y_tr.shape)

print(X_te.shape)
print(y_te.shape)

print(X_va.shape)
print(y_va.shape)

# first index ra akong e print
print(":D \n", X_tr[0], "--->", y_tr[0])
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Bidirectional

model = Sequential()
model.add(Bidirectional((LSTM(50, return_sequences = True, activation='relu')), input_shape=(n_steps, n_features)))
model.add(Bidirectional((LSTM(50)), input_shape=(n_steps, n_features)))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_tr_s, y_tr_s, epochs=60, verbose=0)
```

```python
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
from math import sqrt

def forecast_lstm(model_, batch_size, X_input):
    X_input = X_input.reshape((batch_size, n_steps, n_features))
    #print(X_input)
    yhat = model_.predict(X_input, verbose=0)
    #print(yhat.shape)
    return yhat[0,0]

# @param:
#   model <-- the model 
#   Xval, yval  <-- the validation set
def evaluate_lstm_model(model_, Xval, yval):
    predictions = list()
    for i in range(len(Xval)):
        yhat = forecast_lstm(model_, 1, Xval[i])
        predictions.append(yhat)
        
    # calculate out of sample error
    mse = mean_squared_error(yval, predictions)
    rmse = sqrt(mse)
    print(rmse)
    
    # I cannot dataframe the df_10 because it needs to be a 1d array
    # so i change y_va to 1d
    print(yval.shape)
    Aaa = yval.reshape(yval.shape[0])
    print(yval.shape)
    
    df_10 = {'Expected':Aaa, 'predictions':predictions}
    df_10 = pd.DataFrame(df_10)
    #print(df_10)
    return rmse, df_10["predictions"]
```
